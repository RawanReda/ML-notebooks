# Regression Dataset

Obtaining the dataset: 

```python
train_reg= pd.read_csv('../input/real-estate-price-prediction/Real estate.csv')
```

```python
train_reg.head()
```

```python
train_reg
```

Checking for data that needs cleaning: 

```python
train_reg.info()
```

```python
X=train_reg.drop('Y house price of unit area', axis=1)
y= train_reg['Y house price of unit area']
```

Splitting the dataset: 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Scaling dataset to be used with linear regression and SVR: 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
Sc_X_train=scaler.fit_transform(X_train)
Sc_X_test=scaler.transform(X_test) ## In the test dataset we call transform without fit 
```

```python
lin_reg = LinearRegression(normalize=True)
lin_reg.fit(Sc_X_train,y_train)
```

```python
score_train = lin_reg.score(Sc_X_train, y_train) #Coefficient of determination
score_test =lin_reg.score(Sc_X_test, y_test)

print('train score: {}'.format(score_train)) 
print('test score: {}'.format(score_test))
```

**.score() returns the coefficient of determination, or R², for the data passed. Its maximum is 1. The higher the R² value, the better the fit.** <br/>
**Unlike root mean squared error, where the lower the value, the better the fit.** <br/>
**More on this part will be covered next session, but .score() is R² and not root mean sqaured .** <br/>


```python
lin_reg.intercept_
```

```python
print(lin_reg.coef_)
```

```python
y_pred =lin_reg.predict(Sc_X_test)
y_pred
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
from sklearn.svm import SVR

reg=SVR(kernel='rbf')
reg.fit(X_train, y_train)


svr_score_train = reg.score(X_train, y_train) #Coefficient of determination
svr_score_test =reg.score(X_test, y_test)

print('train score: {}'.format(svr_score_train)) 
print('test score: {}'.format(svr_score_test))
```

# Classification dataset

```python
trainC = pd.read_csv('../input/breast-cancer-wisconsin-data/data.csv')
```

```python
trainC.info()
```

```python
X= trainC.drop(['diagnosis','Unnamed: 32'], axis=1)
y= trainC['diagnosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(max_depth=5)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
